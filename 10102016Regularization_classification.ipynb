## Prepare Data

```python
from sklearn import cross_validation
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import OneHotEncoder
```

```python
array = dataset.values
X = array[:,1:9]
Y = (array[:,10]-2).astype(np.bool)

enc = OneHotEncoder()
enc.fit(X)
X = enc.transform(X).toarray()
```

```python
poly = PolynomialFeatures(interaction_only=True,include_bias=False)
X = poly.fit_transform(X)
```

```python
validation_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

## Tune Models

```python
import numpy as np
np.random.seed(7)
import random
random.seed(7)

from sklearn.grid_search import GridSearchCV
from sklearn.linear_model import SGDClassifier
```

### Tune Lasso 

```python
alphas = 10.0**-np.arange(1,3)
eta0 = 10.0**-np.arange(1,3)
n_iter = np.ceil(10**6 / len(X_train))
param_grid = dict(alpha=alphas,eta0=eta0)
model = SGDClassifier(loss='log', penalty='l1',n_iter=n_iter)
scoring = 'roc_auc' 
num_folds = 10
grid = GridSearchCV(cv=num_folds, estimator=model, param_grid=param_grid,scoring=scoring,n_jobs=-1)
grid.fit(X_train, Y_train)
```

```python
print(grid.best_score_)
print(grid.best_estimator_.alpha)
print(grid.best_estimator_.eta0)
lassoTunedAlpha = grid.best_estimator_.alpha
lassoTunedEta0 = grid.best_estimator_.eta0
```

### Tune Ridge 

```python
alphas = 10.0**-np.arange(1,3)
eta0 = 10.0**-np.arange(1,3)
n_iter = np.ceil(10**6 / len(X_train))
param_grid = dict(alpha=alphas,eta0=eta0)
model = SGDClassifier(loss='log', penalty='l2',n_iter=n_iter)
scoring = 'roc_auc' 
num_folds = 10
grid = GridSearchCV(cv=num_folds, estimator=model, param_grid=param_grid,scoring=scoring,n_jobs=-1)
grid.fit(X_train, Y_train)
```

```python
print(grid.best_score_)
print(grid.best_estimator_.alpha)
print(grid.best_estimator_.eta0)
ridgeTunedAlpha = grid.best_estimator_.alpha
ridgeTunedEta0 = grid.best_estimator_.eta0
```

### Tune Elastic Net

```python
alphas = 10.0**-np.arange(1,3)
eta0 = 10.0**-np.arange(1,3)
l1_ratio = np.arange(0.25,1.0,0.25,dtype=np.float32)
n_iter = np.ceil(10**6 / len(X_train))
param_grid = dict(alpha=alphas, l1_ratio=l1_ratio,eta0=eta0)
model = SGDClassifier(loss='log', penalty='elasticnet',n_iter=n_iter)
scoring = 'roc_auc'
num_folds = 10
grid = GridSearchCV(cv=num_folds, estimator=model, param_grid=param_grid,scoring=scoring,n_jobs=-1)
grid.fit(X_train, Y_train)
```

```python
print(grid.best_score_)
print(grid.best_estimator_.alpha, grid.best_estimator_.l1_ratio, grid.best_estimator_.eta0)
enTunedAlpha = grid.best_estimator_.alpha
enTunedL1Ratio = grid.best_estimator_.l1_ratio
enTunedEta0 = grid.best_estimator_.eta0
```

## Evaluate Performance

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import auc
from sklearn.metrics import roc_curve
```

```python
seed = 1
n_iter = np.ceil(10**6 / len(X_train))
pipelines = []
pipelines.append(('LR', Pipeline([('LR', LogisticRegression())])))
pipelines.append(('LASSO', Pipeline([('LASSO', SGDClassifier(loss='log', penalty='l1',alpha=lassoTunedAlpha,random_state=seed,n_iter=n_iter,eta0=lassoTunedEta0))])))
pipelines.append(('Ridge', Pipeline([('Ridge', SGDClassifier(loss='log', penalty='l2',alpha=ridgeTunedAlpha,random_state=seed,n_iter=n_iter,eta0=ridgeTunedEta0))])))
pipelines.append(('EN', Pipeline([('EN', SGDClassifier(loss='log', penalty='elasticnet',alpha=enTunedAlpha,l1_ratio=enTunedL1Ratio,random_state=seed,n_iter=n_iter,eta0=enTunedEta0))])))
```

```python
# Make predictions on test dataset
results = []
names = []
for name, model in pipelines:
    model.fit(X_train,Y_train)
    fpr, tpr, thresholds = roc_curve(Y_test,model.predict_proba(X_test)[:,1])
    result = auc(fpr,tpr)
    results.append(result)
    names.append(name)
    msg = "%s: %f" % (name, result)
    print(msg)
```
