## Download housing dataset

```python
# CRIM: per capita crime rate by town 
# ZN: proportion of residential land zoned for lots over 25,000 sq.ft. 
# INDUS: proportion of non-retail business acres per town 
# CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise) 
# NOX: nitric oxides concentration (parts per 10 million) 
# RM: average number of rooms per dwelling 
# AGE: proportion of owner-occupied units built prior to 1940 
# DIS: weighted distances to five Boston employment centres 
# RAD: index of accessibility to radial highways 
# TAX: full-value property-tax rate per $10,000
# PTRATIO: pupil-teacher ratio by town 
# B: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town 
# LSTAT: "% lower status of the population"
# MEDV: Median value of owner-occupied homes in $1000's
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
dataset = pd.read_csv(url, delim_whitespace=True, names=names)
```

## Prepare Data

```python
from sklearn import cross_validation
from sklearn.preprocessing import PolynomialFeatures
```

```python
array = dataset.values
X = array[:,0:13]
Y = array[:,13]

poly = PolynomialFeatures(degree=2,interaction_only=False,include_bias=False)
X = poly.fit_transform(X)
```

```python
validation_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(X_train)
rescaledX_train = scaler.transform(X_train)
```

## Tune Models

```python
import numpy as np
np.random.seed(7)
import random
random.seed(7)

from sklearn.grid_search import GridSearchCV
from sklearn.linear_model import SGDRegressor
```

### Tune Lasso 

```python
alphas = 10.0**-np.arange(1,7)
eta0 = 10.0**-np.arange(1,7)
n_iter = np.ceil(10**6 / len(rescaledX_train))
param_grid = dict(alpha=alphas,eta0=eta0)
model = SGDRegressor(loss='squared_loss', penalty='l1',n_iter=n_iter)
scoring = 'mean_squared_error' #r2
num_folds = 10
grid = GridSearchCV(cv=num_folds, estimator=model, param_grid=param_grid,scoring=scoring,n_jobs=-1)
grid.fit(rescaledX_train, Y_train)
```

```python
print(grid.best_score_)
print(grid.best_estimator_.alpha)
print(grid.best_estimator_.eta0)
lassoTunedAlpha = grid.best_estimator_.alpha
lassoTunedEta0 = grid.best_estimator_.eta0
```

### Tune Ridge 

```python
alphas = 10.0**-np.arange(1,7)
eta0 = 10.0**-np.arange(1,7)
n_iter = np.ceil(10**6 / len(rescaledX_train))
param_grid = dict(alpha=alphas,eta0=eta0)
model = SGDRegressor(loss='squared_loss', penalty='l2',n_iter=n_iter)
scoring = 'mean_squared_error' #r2
num_folds = 10
grid = GridSearchCV(cv=num_folds, estimator=model, param_grid=param_grid,scoring=scoring,n_jobs=-1)
grid.fit(rescaledX_train, Y_train)
```

```python
print(grid.best_score_)
print(grid.best_estimator_.alpha)
print(grid.best_estimator_.eta0)
ridgeTunedAlpha = grid.best_estimator_.alpha
ridgeTunedEta0 = grid.best_estimator_.eta0
```

### Tune Elastic Net

```python
alphas = 10.0**-np.arange(1,3)
eta0 = 10.0**-np.arange(1,3)
l1_ratio = np.arange(0.25,1.0,0.25,dtype=np.float32)
n_iter = np.ceil(10**6 / len(rescaledX_train))
param_grid = dict(alpha=alphas, l1_ratio=l1_ratio,eta0=eta0)
model = SGDRegressor(loss='squared_loss', penalty='elasticnet',n_iter=n_iter)
scoring = 'mean_squared_error' #r2
num_folds = 10
grid = GridSearchCV(cv=num_folds, estimator=model, param_grid=param_grid,scoring=scoring,n_jobs=-1)
grid.fit(rescaledX_train, Y_train)
```

```python
print(grid.best_score_)
print(grid.best_estimator_.alpha, grid.best_estimator_.l1_ratio, grid.best_estimator_.eta0)
enTunedAlpha = grid.best_estimator_.alpha
enTunedL1Ratio = grid.best_estimator_.l1_ratio
enTunedEta0 = grid.best_estimator_.eta0
```

## Evaluate Performance

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
seed = 1
n_iter = np.ceil(10**6 / len(X_train))
pipelines = []
pipelines.append(('LR', Pipeline([('Scaler', StandardScaler()),('LR', LinearRegression())])))
pipelines.append(('LASSO', Pipeline([('Scaler', StandardScaler()),('LASSO', SGDRegressor(loss='squared_loss', penalty='l1',alpha=lassoTunedAlpha,random_state=seed,n_iter=n_iter,eta0=lassoTunedEta0))])))
pipelines.append(('Ridge', Pipeline([('Scaler', StandardScaler()),('Ridge', SGDRegressor(loss='squared_loss', penalty='l2',alpha=ridgeTunedAlpha,random_state=seed,n_iter=n_iter,eta0=ridgeTunedEta0))])))
pipelines.append(('EN', Pipeline([('Scaler', StandardScaler()),('EN', SGDRegressor(loss='squared_loss', penalty='elasticnet',alpha=enTunedAlpha,l1_ratio=enTunedL1Ratio,random_state=seed,n_iter=n_iter,eta0=enTunedEta0))])))
```

```python
# Make predictions on test dataset
results = []
names = []
for name, model in pipelines:
    model.fit(X_train,Y_train)
    result = mean_squared_error(Y_test,model.predict(X_test))
    results.append(result)
    names.append(name)
    msg = "%s: %f" % (name, result)
    print(msg)
```
